### Answer of Problem 4.2

```r
# Convert to date format
data["DATE"] <- as.Date(data[["DATE"]])
# Get the column names
cols <-colnames(data)
date <- data['DATE']
```

```r
leng <- (length(cols)-1)/2
cap_data <- data.frame(Date = date)
res_data <- data.frame(Date = date)

for(i in 1:leng){
    k <- 2*i
    j <- 2*i + 1
    res_data[cols[j]]<-data[j]
    cap_data[cols[k]]<- data[k]
}
```

```r
tail(data)
```

```r
# Define a function to comupte covariance sigma
# Last day: 2016-12-31
Big_Sigma <- function(shrinkage, res_data, t ="2016-12-31"){
    df_data <- res_data[res_data[["DATE"]] <= t, ]
    vols <- c()
    
    for(i in 2:16){vols <- c(vols, sd(df_data[[i]]))}

    S <- diag(vols)
    R <- cor(subset(df_data, select = -c(DATE)))

    size <- dim(R)[1]
    Ident <- diag(size)
    Conv <- S %*% (shrinkage*R + (1-shrinkage)*Ident) %*% S # Covariance matrix formula
    return(list(kappa = kappa(Conv,exact = TRUE), Conv = Conv))
}
```

```r
# Create shrinkage list
shrk_list <- seq(0.1, 1, 0.005)
condition_num <- sapply(shrk_list, Big_Sigma, res_data = res_data)["kappa",]

# Plot the condition number vs shrinkage
options(repr.plot.height=5, repr.plot.width=8)
plot(shrk_list, condition_num, xlab = "Shrinkage", 
     ylab = "Condition Number", type='l', col='blue')
```

- From the figure above, we can conclude that 0.8 is a good approximate of shrinkage to achieve a well-conditioned convariance matrix

### Answer of Problem 4.3

```r
# Seperate the dataset as given condition based on weights
df_Scandres <- c('country.nor.totret','country.swe.totret', 'country.dnk.totret', 'country.fin.totret')
# Cap
df_Scandcap <- c('country.nor.previous.usdcap','country.swe.previous.usdcap', 'country.dnk.previous.usdcap', 'country.fin.previous.usdcap')
```

```r
# Define a function to get the weights (P)
Get_P <- function(capdata, t){
    capcols <- colnames(capdata)
    n <- length(capcols)
    nonScandcapdata <- capdata[,-which(names(capdata) %in% df_Scandcap)]
    sum <- sum(nonScandcapdata[nonScandcapdata[["DATE"]]==t,][-1])
    weights <- list()
    for(i in 2:n){
        if(any(capcols[i] %in% df_Scandcap)){
            weights[capcols[i]] <- 0.25
        }
        else{
            weights[capcols[i]] <- -(capdata[capdata[["DATE"]]==t,i]/sum)
        }
    }
    return(weights)
}
```

### Answer of Problem 4.4

```r
# Define a function to compute Black-Literman Portfolio
pf_BL <- function(capdata, resdata, t, omega, q){
    date <- capdata[["DATE"]]
    index <- which(date==t)
    tprev <- date[index-1] # avoid lookback bias
    
    # Set shrinkage as 0.5
    # Compute covariance sigma
    Sigma <- Big_Sigma(0.5, resdata, tprev)$Conv
    
    capdata_t <- capdata[capdata[["DATE"]]==tprev,]
    sum <- sum(capdata_t[-1])
    h_eq <- c()
    for(i in 2:16){
        temp <- capdata_t[i]/sum
        h_eq <- c(h_eq, temp)
    }
    h_eq <- as.vector(unlist(h_eq))
    
    riskaver <- 1  # Given kappa as 1
    tau <- 0.01    # Given tau as 0.01
    
    Pi <- riskaver *(1+tau)*(Sigma %*% h_eq)
    P <- as.matrix(unlist(Get_P(capdata, t)))
    C <- tau * Sigma
    
    
    H_inv <- solve(P%*%t(P)/omega + solve(C))
    h_star <- solve(H_inv + Sigma) %*% H_inv %*%(P*q/omega + solve(C)%*%Pi)/riskaver

    return(list(h_star=unlist(h_star), h_eq=h_eq))
    
    
}
```

```r
# Compute the portfolio
# Condition from Problem 4.3: omega and q
omega <- 0.015 
q <- 0.01
res_BL <- pf_BL(cap_data,res_data,"2016-12-31",omega=omega,q=q)
```

```r
# Get the hstar and heq
hstar <- res_BL$h_star
heq <- res_BL$h_eq
```

### Answer of Problem 4.5

```r
time <- res_data[['DATE']]

# res of two returns
res_star <- c()
res_eq <- c()

omega <-0.015
q <- 0.01
for(i in 51:202){
    t <- as.Date(unlist(time[i]))
    res <- pf_BL(cap_data, res_data, t,omega,q)
    hstar <- res$h_star
    heq <- res$h_eq
    # Return in month t
    r_t <- as.vector(unlist(res_data[res_data[["DATE"]]==time[i+1],][-1]))
    
    star_ret <- t(hstar)%*%r_t
    eq_ret <- t(heq)%*%r_t
    
    res_star<-c(res_star, star_ret)
    res_eq<- c(res_eq, eq_ret)
}
```

```r
# Compute the cumulative sum of star and equlibirum
res_Q5 <- data.frame(DATE = time[51:202], Optimal_Portfolio = cumsum(res_star), Equilibrium_Portfolio = cumsum(res_eq))
```

```r
# Reshape the dataframe
res_Q5_melt <- melt(res_Q5, id.vars = "DATE")
```

```r
head(res_Q5_melt)
```

```r
# Plot the two time series vs time
options(repr.plot.height=5, repr.plot.width=8)
ggplot(res_Q5_melt, aes(x = DATE, y = value, color = variable) )+
  theme_bw() +
  geom_point()+
  labs(x="Date", y = "Cumulative sum of portfolio returns")+
  geom_line()
```
